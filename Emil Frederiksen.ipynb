```python
# Rename columns
xg_df = xg_df.rename(columns={'321': 'xg', '322': 'post_shot_xg'})

# Convert to float
xg_df['xg'] = xg_df['xg'].astype(float)
xg_df['post_shot_xg'] = xg_df['post_shot_xg'].astype(float)

# Group by playerName and sum
xg_df = xg_df.groupby('playerName').sum()[['xg', 'post_shot_xg']]

# Calculate xG difference
xg_df['xg_diff'] = xg_df['post_shot_xg'] - xg_df['xg']

# Rank all metrics (higher is better)
xg_df['rank_xg'] = xg_df['xg'].rank(method='min', ascending=False)
xg_df['rank_post_shot_xg'] = xg_df['post_shot_xg'].rank(method='min', ascending=False)
xg_df['rank_xg_diff'] = xg_df['xg_diff'].rank(method='min', ascending=False)

# Optional: sort by a specific rank
xg_df = xg_df.sort_values(by='rank_xg')
xg_df = xg_df.reset_index()
print(xg_df)
xg_df_emil = xg_df[xg_df['playerName'] == 'E. Frederiksen']
print(xg_df_emil)

big_chance_df = matchstats_df.groupby('player_matchName').sum()[['penAreaEntries', 'successfulFinalThirdPasses','bigChanceCreated']].reset_index()
big_chance_df['rank_penAreaEntries_rank'] = big_chance_df['penAreaEntries'].rank(method='min', ascending=False)
big_chance_df['rank_successfulFinalThirdPasses'] = big_chance_df['successfulFinalThirdPasses'].rank(method='min', ascending=False)
big_chance_df['rank_bigChanceCreated'] = big_chance_df['bigChanceCreated'].rank(method='min', ascending=False)
big_chance_df = big_chance_df[big_chance_df['player_matchName']== 'E. Frederiksen']
print(big_chance_df)
```
